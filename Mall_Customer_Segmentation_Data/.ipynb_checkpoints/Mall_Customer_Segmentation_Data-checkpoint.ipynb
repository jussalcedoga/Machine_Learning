# Segmentation using Annual Income and Spending Score

```python
X1 = data[['Annual Income (k$)' , 'Spending Score (1-100)']].iloc[: , :].values
squared_distance = []
for n in range(1 , 11):
    algorithm = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
    algorithm.fit(X1)
    squared_distance.append(algorithm.inertia_)
```

```python
plt.figure(1 , figsize = (15 , 7))
plt.plot(numpy.arange(1, 11), squared_distance, "-o", ms = 10)
plt.xlabel(r"$\rm{Number \ of \ Clusters}$", fontsize = 20)
plt.ylabel(r"$\rm{Squared \ Error}$", fontsize = 20)
plt.show()
```

From this, one can estimate that the needed amount of clusters for this particular segmentation, corresponds to 5. It is even clearer from the following plot.

```python
algorithm = (KMeans(n_clusters = 5 ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
algorithm.fit(X1)
labels2 = algorithm.labels_
centroids2 = algorithm.cluster_centers_

h = 0.02
x_min, x_max = X1[:, 0].min() - 1, X1[:, 0].max() + 1
y_min, y_max = X1[:, 1].min() - 1, X1[:, 1].max() + 1
xx, yy = numpy.meshgrid(numpy.arange(x_min, x_max, h), numpy.arange(y_min, y_max, h))
Z2 = algorithm.predict(numpy.c_[xx.ravel(), yy.ravel()]) 

plt.figure(1 , figsize = (15 , 7) )
plt.clf()
Z2 = Z2.reshape(xx.shape)
plt.imshow(Z2 , interpolation='nearest', 
           extent=(xx.min(), xx.max(), yy.min(), yy.max()),
           cmap = plt.cm.Pastel2, aspect = 'auto', origin='lower')

plt.scatter( x = 'Annual Income (k$)' ,y = 'Spending Score (1-100)' , data = data , c = labels2 , 
            s = 200 )
plt.scatter(x = centroids2[: , 0] , y =  centroids2[: , 1] , s = 300 , c = 'red' , alpha = 0.5)
plt.ylabel('Spending Score (1-100)') , plt.xlabel('Annual Income (k$)')
plt.show()
```

From this plot, it can be seen that 5 is a good amount of clusters for grouping the data.

# Segmentation using three Age, Spending Score and Annual Income

```python
X4 = data[['Age' , 'Annual Income (k$)' ,'Spending Score (1-100)']].iloc[: , :].values
squared_distance = []
for n in range(1 , 11):
    algorithm = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
    algorithm.fit(X4)
    squared_distance.append(algorithm.inertia_)
```

```python
plt.figure(1 , figsize = (15 , 7))
plt.plot(numpy.arange(1, 11), squared_distance, "-o", ms = 10)
plt.xlabel(r"$\rm{Number \ of \ Clusters}$", fontsize = 20)
plt.ylabel(r"$\rm{Squared \ Error}$", fontsize = 20)
plt.show()
```

```python
algorithm = (KMeans(n_clusters = 6 ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 111  , algorithm='elkan') )
algorithm.fit(X4)
labels3 = algorithm.labels_
centroids3 = algorithm.cluster_centers_
```

```python
data['label3'] =  labels3
trace1 = go.Scatter3d(
    x= data['Age'],
    y= data['Annual Income (k$)'],
    z= data['Spending Score (1-100)'],
    mode='markers',
     marker=dict(
        color = data['label3'], 
        size= 20,
        line=dict(
            color= data['label3'],
            width= 12
        ),
        opacity=0.8
     )
)
data_ = [trace1]
layout = go.Layout(
#     margin=dict(
#         l=0,
#         r=0,
#         b=0,
#         t=0
#     )
    title= 'Clusters',
    scene = dict(
            xaxis = dict(title  = 'Age'),
            yaxis = dict(title  = 'Spending Score'),
            zaxis = dict(title  = 'Annual Income')
        )
)
fig = go.Figure(data=data_, layout=layout)
py.offline.iplot(fig)
```

From this plot, it can be seen the 6 clusters which were needed to group the data according to the "elbow" analysis performed above. A total amount of 6 clusters can group the data in a reliable manner. These cluster were identified or estimated in terms of the three main variables which are age, annual income, as well as Spending Score. One could have thougth from the beggining that the gender would have been such an important parameter to take into account when thinking on customer segmentation, but as the analysis showed up, the gender was not an indicative of such a segmentation for this particular dataset.
